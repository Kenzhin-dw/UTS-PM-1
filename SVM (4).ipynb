```python
import numpy as np
import pandas as pd
```

Membaca data

```python
df = pd.read_csv("K03_counter_strike_2.CSV")
df
```

Menampilkan Nilai Unik

```python
df['map'].unique()
```

Informasi Data

```python
df.info()
```

Encoding Kolom map

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['map'] = le.fit_transform(df['map'])
df
```

```python
df['map'].unique()
```

Encoding Kolom round_winner

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['round_winner'] = le.fit_transform(df['round_winner'])
df
```

```python
df['round_winner'].unique()
```

Menyiapkan Data Fitur dan Target

```python
x = np.asarray(df.drop(['round_winner', 'map'], axis = 1), dtype = np.float32)
y = np.asarray(df['map'], dtype = np.int32)
```

```python
x
```

Split Data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

model = SVC()
model.fit(x_train, y_train)

y_pred = model.predict(x_train)
print(accuracy_score(y_train, y_pred))

y_pred_test = model.predict(x_test)
print(accuracy_score(y_test, y_pred_test))
```

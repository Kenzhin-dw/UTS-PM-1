Import library

```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression 
```

Membaca Data

```python
df = pd.read_csv ("R03_excercises.csv")
df
```

Memeriksa Missing Values

```python
df.isna().sum()
```

Menghitung Korelasi:

```python
df.corr()
```

```python
Mengganti Nama Kolom
```

```python
df.columns = ['x', 'y']
df
```

Menyiapkan Data Latih

```python
x_train = df['x'].values[:,np.newaxis]
y_train = df['y'].values

lm = LinearRegression()
lm.fit(x_train, y_train)
```

Melatih Model Regresi Linear:

```python
x_test = [[4.0], [2.5], [1.9], [0.4], [3.0]]
p = lm.predict(x_test)
print(p)
```

Koefisien dan Intersep:

```python
print('Coefficient : ' + str(lm.coef_))
print('Intercept : ' + str(lm.intercept_))
```

```python
y_asli = [0.8, 1.7, 0.4]
y_hasil_prediksi = lm.predict([[1.9], [3.4], [1.0]])
print(y_hasil_prediksi)
```

```python
pb = lm.predict(x_train)
dfc = pd.DataFrame({'x' : df['x'], 'y' : pb})
plt.scatter(df['x'], df['y'])
plt.plot(dfc['x'], dfc['y'], color = 'red', linewidth = 1) 
plt.xlabel('Daily Exercise Time')
plt.ylabel('Weight Loss')
plt.show
```
